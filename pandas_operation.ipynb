```python
def getgrade(marks):
    p=(marks/200)*100   #percentage 
    if p>=75:
        return "A"
    elif (p<75 and p>=60):
        return "B"
    elif (p<60 and p>=45):
        return "C"
    elif (p<45 and p>=30):
        return "D"
    else :
        return "F"
df["total"].apply(getgrade)
```
